```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import time

# trends=TrendReq
```

```python
# Instantiate the TrendReq class
trends = TrendReq()

# Introduce a delay before making the request
time.sleep(5)  # You can adjust this value based on your needs

# Use kw_list and specify a time frame
trends.build_payload(kw_list=["Blackpink"], timeframe='2022-01-01 2023-01-01')

# Introduce a delay before making the second request
time.sleep(1)

# Get interest by region
data = trends.interest_by_region()
data = data.sort_values(by="Blackpink", ascending=False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x="geoName",y="Blackpink",figsize=(20,15),kind="line")

plt.style.use('fivethirtyeight')
plt.show





```

```python
# Get interest by region
data = trends.interest_by_region()
data = data.sort_values(by="Blackpink", ascending=False)
data = data.head(10)

# Reset index for plotting
data_reset = data.reset_index()

# Create a bar chart
plt.figure(figsize=(20, 15))
plt.bar(data_reset["geoName"], data_reset["Blackpink"], color='blue', label='Interest Level')

# Create a line plot
plt.plot(data_reset["geoName"], data_reset["Blackpink"], color='red', marker='o', label='Interest Level')

plt.style.use('fivethirtyeight')
plt.xlabel('Region')
plt.ylabel('Interest Level')
plt.title('Interest in Blackpink by Region')
plt.legend()
plt.show()
```
